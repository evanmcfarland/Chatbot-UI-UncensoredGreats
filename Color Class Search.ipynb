```python
import os
import re

# Define the path to your project directory
project_directory = "."

# Define the regex pattern to match color classes
color_pattern = re.compile(r"(bg|text|border|hover:bg|hover:text|dark:bg|focus:border|dark:border|dark:text|focus:text)-[A-Za-z0-9-/]+")

color_set = set()

# Walk through the project directory
for root, _, files in os.walk(project_directory):
    for file in files:
        if file.endswith((".js", ".jsx", ".ts", ".tsx")):
            with open(os.path.join(root, file), "r", encoding="utf-8") as f:
                content = f.read()
                matches = color_pattern.findall(content)
                for match in matches:
                    color_set.add(match)

# Print the unique color classes
for color in color_set:
    print(color)
```
